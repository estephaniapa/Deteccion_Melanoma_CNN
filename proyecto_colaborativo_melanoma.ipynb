### Melanoma

***Biología y etiología del melanoma***

<t align="justify">

El melanoma es un tipo de cáncer de piel que se origina en las células productoras de pigmento llamadas melanocitos. Estas células son responsables de producir melanina, el pigmento que da color a la piel, el cabello y los ojos. El melanoma puede desarrollarse tanto en la piel expuesta al sol como en áreas no expuestas, como las membranas mucosas y los ojos (*OMS, 2021*).

***Relevancia del diagnóstico preciso del melanoma***

<t align="justify">

El diagnóstico preciso del melanoma es fundamental debido a su agresividad y capacidad metastásica. La detección temprana y el tratamiento oportuno pueden marcar la diferencia en la supervivencia de los pacientes. Si se detecta en etapas tempranas, el melanoma tiene una alta tasa de curación, pero su progresión puede ser rápida si no se trata adecuadamente (*Mayo clinic, 2021*).

***Estadísticas sobre el melanoma***

<t align="justify">


El melanoma es una enfermedad que ha experimentado un aumento significativo en su incidencia en las últimas décadas. Según la Organización Mundial de la Salud (*OMS, 2021*), se estima que en el año 2020 se diagnosticaron aproximadamente 287,723 nuevos casos de melanoma en todo el mundo. Además, se estima que el melanoma causó alrededor de 60,712 muertes en el mismo año.


***Dificultades de diagnóstico con métodos convencionales***

<t align="justify">

El diagnóstico del melanoma puede presentar desafíos debido a su variabilidad clínica y la similitud de sus características con otras lesiones cutáneas benignas. Los métodos de diagnóstico convencionales, como la inspección visual por parte de médicos especializados y la dermatoscopia, pueden depender en gran medida de la experiencia y la subjetividad del médico, lo que puede llevar a variaciones en la precisión diagnóstica (*Mayo Clinic, 2021; OMS, 2021*).

***El potencial de las redes neuronales convolucionales en el diagnóstico del melanoma***

<t align="justify">

Las redes neuronales convolucionales (RNC) han demostrado un gran potencial en el campo de la visión por computadora y el análisis de imágenes médicas. Estas redes son capaces de aprender características discriminativas directamente de las imágenes, lo que las convierte en una herramienta prometedora para la detección automática y precisa del melanoma. Al entrenar una RNC con una base de datos de imágenes clasificadas, se puede lograr que la red aprenda a distinguir entre lesiones melanocíticas malignas y benignas.

#### La Red Neuronal Convolucional 

***Descripción general de las redes neuronales***

<t align="justify">

Las redes neuronales artificiales (RNA) son modelos computacionales inspirados en el funcionamiento del cerebro humano, diseñados para aprender y procesar información de manera similar a las redes de neuronas biológicas. Estas redes están compuestas por unidades básicas llamadas neuronas artificiales o nodos, que se organizan en capas y están interconectadas mediante conexiones ponderadas (*Rumelhart et al., 1986*).

Las RNA tienen una amplia gama de aplicaciones en diversos campos, incluyendo reconocimiento de voz, visión por computadora, procesamiento del lenguaje natural, pronóstico de series temporales y diagnóstico médico, entre otros. Su capacidad para aprender a partir de ejemplos y detectar patrones complejos en los datos ha hecho que sean una herramienta poderosa en la inteligencia artificial y el aprendizaje automático.

El proceso de entrenamiento de una RNA implica la presentación de un conjunto de datos de entrenamiento, que consiste en pares de entradas y salidas deseadas. Durante el entrenamiento, la red ajusta los pesos de las conexiones entre las neuronas para minimizar una función de pérdida que mide la discrepancia entre las salidas predichas y las salidas deseadas. Esto se logra mediante técnicas de optimización, como el descenso del gradiente, que buscan encontrar los valores óptimos de los pesos para que la red pueda hacer predicciones precisas (*Rumelhart et al., 1986*).

A medida que una RNA se entrena con más ejemplos, se espera que mejore su capacidad para generalizar y realizar predicciones precisas en datos no vistos anteriormente. El proceso de entrenamiento puede requerir ajustes en los hiperparámetros de la red, como la tasa de aprendizaje, el número de capas ocultas y el tamaño de las capas, para lograr un equilibrio entre el ajuste preciso a los datos de entrenamiento y la capacidad de generalización a nuevos datos.

***Redes neuronales convencionales y sus limitaciones***

<t align="justify">

Las redes neuronales convencionales, como el perceptrón multicapa, han sido ampliamente utilizadas en una variedad de tareas de aprendizaje automático. Estas redes están compuestas por múltiples capas de neuronas interconectadas, incluyendo una capa de entrada, una o varias capas ocultas y una capa de salida. Cada neurona en una capa se conecta con todas las neuronas de la capa siguiente mediante conexiones ponderadas (*LeCun, Bengio, & Hinton, 2015*).

Estas redes neuronales convencionales son capaces de aprender y resolver una amplia gama de problemas, desde clasificación de imágenes hasta predicción de series temporales. Durante el entrenamiento, los pesos de las conexiones entre las neuronas se ajustan iterativamente utilizando técnicas como el descenso del gradiente y la retropropagación del error.

A pesar de su eficacia en muchas tareas, las redes neuronales convencionales tienen algunas limitaciones. Una de las principales limitaciones es su rendimiento en tareas que involucran el procesamiento de imágenes y otros tipos de datos con estructura espacial. Estos tipos de datos presentan patrones y características que pueden ser difíciles de capturar y modelar eficientemente mediante conexiones densamente interconectadas (*Krizhevsky, Sutskever, & Hinton, 2012*).

Es en este contexto que las redes neuronales convolucionales (RNC) han surgido como una solución efectiva. Las RNC se destacan por su capacidad para procesar datos con estructura espacial, como imágenes, de manera más eficiente que las redes neuronales convencionales. A través de su arquitectura especializada, las RNC pueden capturar patrones locales y extraer características relevantes en diferentes niveles de abstracción.

En comparación con las redes neuronales convencionales, las RNC ofrecen varias ventajas, como la reducción del número de parámetros, la compartición de pesos y la invariancia a transformaciones locales. Estas características hacen que las RNC sean especialmente adecuadas para tareas de visión por computadora, reconocimiento de objetos y detección de patrones en imágenes (*LeCun et al., 2015; Krizhevsky et al., 2012*).

En las secciones siguientes, profundizaremos en los componentes fundamentales de una RNC y cómo abordan las limitaciones de las redes neuronales convencionales en el procesamiento de imágenes.

***Redes neuronales convolucionales: características y aplicaciones en procesamiento de imágenes***

<t align="justify">


Las redes neuronales convolucionales o RNC son una variante especializada de las redes neuronales artificiales que han demostrado un gran éxito en el procesamiento de imágenes. Estas redes se han convertido en un componente fundamental en aplicaciones de visión por computadora, como reconocimiento de objetos, detección de patrones y clasificación de imágenes.

A diferencia de las redes neuronales convencionales, las RNC están diseñadas específicamente para aprovechar las características espaciales presentes en las imágenes. Utilizan capas convolucionales que aplican filtros (kernels) a las regiones locales de la imagen para extraer características relevantes. Esta capacidad de procesamiento localizado permite a las RNC capturar patrones visuales, como bordes, texturas y formas, en diferentes niveles de abstracción (*LeCun et al., 1998*).

Además de las capas convolucionales, las RNC también incorporan otras capas clave, como capas de activación (generalmente utilizando la función ReLU), capas de pooling y capas completamente conectadas. Estas capas se combinan para formar una arquitectura jerárquica que extrae características cada vez más abstractas a medida que se profundiza en la red (*Goodfellow, Bengio, & Courville, 2016*).

Una de las principales aplicaciones de las RNC es el procesamiento de imágenes. Estas redes pueden aprender automáticamente a reconocer objetos, detectar características específicas en imágenes y clasificar imágenes en diferentes categorías. Su capacidad para capturar y modelar características espaciales en las imágenes las hace especialmente adecuadas para tareas de reconocimiento de objetos y clasificación de imágenes (*LeCun et al., 1998*).

En las secciones siguientes, profundizaremos en los componentes fundamentales de una RNC, incluyendo la convolución y la activación (ReLU), el pooling, la capa de flatten y la capa completamente conectada. Exploraremos en detalle cómo estas operaciones contribuyen al procesamiento de imágenes y cómo se combinan para formar una red neuronal convolucional completa.

***Convolución y activación (ReLU) en redes neuronales convolucionales***

<t align="justify">

Las RNC se caracterizan por su capacidad para procesar imágenes de manera eficiente y capturar características importantes mediante el uso de convoluciones y funciones de activación (*LeCun, Bengio, & Hinton, 2015*).

La convolución es una operación fundamental en las RNC que permite capturar patrones locales en las imágenes. En esta operación, se utiliza un kernel, que es una pequeña ventana deslizante, para realizar una multiplicación elemento por elemento entre los valores de los píxeles de la imagen y los valores del kernel (*Goodfellow, Bengio, & Courville, 2016*). Esto se puede expresar matemáticamente de la siguiente manera:



\begin{eqnarray} 
  \mathbf{C}(i, j) = \sum_{m}\sum_{n} \mathbf{I}(i+m, j+n) \cdot \mathbf{K}(m, n)
\end{eqnarray}


<t align="justify">

donde $\mathbf{C}(i, j)$ es el valor resultante de la convolución en la posición $(i, j)$, $\mathbf{I}(i+m, j+n)$ es el valor del píxel de la imagen en la posición $(i+m, j+n)$, y $\mathbf{K}(m, n)$ es el valor del kernel en la posición $(m, n)$.

La convolución se aplica en toda la imagen deslizando el kernel sobre ella, generando así una representación convolucionada de la imagen que resalta características locales. Esta operación permite a la red neuronal convolucional detectar bordes, texturas y otros patrones visuales importantes en diferentes regiones de la imagen.

Después de la convolución, se aplica la función de activación Rectificadora linear (o ReLU por sus siglas en inglés) para introducir no linealidad en la red (*Krizhevsky et al., 2012*). La función ReLU se define de la siguiente manera:


\begin{eqnarray} 
f(x) = \max(0, x)
\end{eqnarray}


<t align="justify">

donde $x$ es el valor de entrada. La función ReLU mapea los valores negativos a cero y mantiene los valores positivos sin cambios. Esta función de activación juega un papel crucial en la capacidad de la red para aprender representaciones más complejas y discriminativas (*Krizhevsky et al., 2012*).

La combinación de la convolución y la activación ReLU permite a las RNC capturar y resaltar características importantes en las imágenes de manera eficiente y no lineal (*LeCun et al., 2015*).

***Pooling en Redes Neuronales Convolucionales***

<t align="justify">

Después de la etapa de convolución y activación en las RNC, se aplica una operación conocida como "pooling". El pooling es una técnica que reduce la dimensionalidad espacial de la representación convolucionada, preservando las características más importantes (*Krizhevsky, Sutskever, & Hinton, 2012*).

El tipo más común de pooling es el max pooling, donde se divide la imagen en regiones no superpuestas y se toma el valor máximo de cada región. Esta operación se puede expresar de la siguiente manera:


\begin{eqnarray} 
\mathbf{P}(i, j) = \max_{m,n} \left( \mathbf{C}(i \cdot s + m, j \cdot s + n) \right)
\end{eqnarray}

<t align="justify">

donde $\mathbf{P}(i, j)$ es el valor resultante del pooling en la posición $(i, j)$, $\mathbf{C}(i \cdot s + m, j \cdot s + n)$ es el valor en la representación convolucionada en la posición $(i \cdot s + m, j \cdot s + n)$, y $s$ es el tamaño del pool o filtro (*Krizhevsky et al., 2012*).

El pooling se utiliza para reducir el tamaño espacial de la representación, lo que a su vez reduce la cantidad de parámetros y operaciones en la red. Además, el pooling proporciona invariancia a pequeñas traslaciones y deformaciones en las características detectadas, lo que puede mejorar la capacidad de generalización de la red (*Krizhevsky et al., 2012*).

Otro tipo de pooling comúnmente utilizado es el average pooling, que toma el promedio de los valores en lugar del máximo. Sin embargo, el max pooling ha demostrado ser más efectivo en la práctica, ya que preserva las características más relevantes y proporciona una mejor invariancia a pequeñas perturbaciones (*Krizhevsky et al., 2012*).

El proceso de pooling se repite en varias capas de la red, reduciendo gradualmente la dimensionalidad espacial y extrayendo características cada vez más abstractas. Esto permite a la red focalizarse en las características más importantes y generalizar mejor a diferentes instancias de una misma clase (*Krizhevsky et al., 2012*).



***Reducción de dimensionalidad***

<t align="justify">

Después del proceso de pooling, la representación resultante se somete a una reducción de dimensionalidad en las RNC. Esta etapa tiene como objetivo transformar la representación multidimensional obtenida en una forma más manejable y adecuada para la clasificación final.

La reducción de dimensionalidad se logra mediante el proceso de aplanamiento o flatten, donde la representación multidimensional se convierte en un vector de una única dimensión. Este paso garantiza que toda la información relevante extraída de las capas convolucionales y de pooling se conserve y pueda ser utilizada por las capas posteriores de la RNC (*LeCun et al., 1998*).

Este proceso se representa matemáticamente de la siguiente manera:



\begin{eqnarray} 
  \text{Vector unidimensional} = \text{aplanar}(\text{Representación multidimensional})
\end{eqnarray}


Donde la función $\text{aplanar}(\cdot)$ convierte la representación multidimensional en un vector lineal (*Goodfellow et al., 2016*).





***Capa Completamente Conectada***

<t align="justify">

Después de la reducción de dimensionalidad, la representación unidimensional resultante se alimenta a una capa completamente conectada en las RNC. Esta capa tiene como objetivo finalizar el proceso de clasificación y tomar decisiones basadas en la información extraída de las capas anteriores.

La capa completamente conectada está compuesta por neuronas que están conectadas a todas las neuronas de la capa anterior. Cada neurona realiza una combinación lineal de las entradas y las somete a una función de activación no lineal, como la función sigmoide o la función ReLU, para introducir no linealidades en el modelo (*Goodfellow et al., 2016*).

Matemáticamente, la operación realizada por una neurona en la capa completamente conectada se puede expresar como:


\begin{eqnarray} 
  \mathbf{a} = f\left(\sum_{i=1}^{n} w_i \cdot x_i + b\right)
\end{eqnarray}

<t align="justify">

Donde $\mathbf{a}$ es la salida de la neurona, $f$ es la función de activación, $w_i$ son los pesos de conexión, $x_i$ son las entradas correspondientes y $b$ es el sesgo o bias (*Goodfellow et al., 2016*).

La capa completamente conectada puede constar de múltiples neuronas, y la cantidad de neuronas y las funciones de activación utilizadas dependen de la arquitectura específica de la RNC y la tarea de clasificación que se esté abordando (*LeCun et al., 1998*).

***Optimización de la red neuronal convolucional***

<t align="justify">

La optimización de una red neuronal es un proceso fundamental que busca ajustar los parámetros de la red para que pueda realizar una clasificación precisa. En el caso de las RNC, se optimizan los parámetros tanto en la parte previa a la red completamente conectada como en la propia red completamente conectada.

En la parte previa a la red completamente conectada, se utilizan técnicas de optimización para ajustar los pesos de las conexiones en las capas convolucionales y de pooling. Un método popularmente utilizado en esta etapa es el algoritmo Adam (*Kingma & Ba, 2015*). Adam combina las ventajas del algoritmo de descenso de gradiente estocástico (o SGD por sus siglas en inglés) con una adaptación del coeficiente de aprendizaje en función del momento y la estimación del segundo momento del gradiente.

Matemáticamente, el algoritmo Adam actualiza los pesos ww de la red en cada iteración utilizando las siguientes ecuaciones:


\begin{eqnarray} 
  m = \beta_1 \cdot m + (1 - \beta_1) \cdot \nabla_w J(w)
\end{eqnarray}

\begin{eqnarray} 
  v = \beta_2 \cdot v + (1 - \beta_2) \cdot (\nabla_w J(w))^2
\end{eqnarray}

\begin{eqnarray} 
  \hat{m} = \frac{m}{1 - \beta_1^t}
\end{eqnarray}
 
\begin{eqnarray} 
  \hat{v} = \frac{v}{1 - \beta_2^t}
\end{eqnarray}

\begin{eqnarray} 
  w = w - \alpha \cdot \frac{\hat{m}}{\sqrt{\hat{v}} + \epsilon}
\end{eqnarray}


<t align="justify">

Donde $\nabla_w$ es el gradiente de la función de pérdida $J(w)$ con respecto a los pesos $w$, $\beta_1$ y $\beta_2$ son hiperparámetros que controlan la ponderación del momento y el segundo momento del gradiente, respectivamente, $m$ y $v$ son estimaciones del primer momento (media) y segundo momento (varianza) del gradiente, $t$ es el contador de iteraciones, $\hat{m}$ y $\hat{v}$ son las versiones corregidas de los momentos, $\alpha$ es el coeficiente de aprendizaje y $\epsilon$ es una pequeña constante para evitar divisiones por cero.

En la capa completamente conectada de la RNC, se utiliza la función de activación softmax para obtener una distribución de probabilidad sobre las posibles clases de salida. La función softmax se define matemáticamente de la siguiente manera:


\begin{eqnarray} 
  \text{softmax}(\mathbf{z})_j = \frac{{e^{z_j}}}{{\sum_{k=1}^{K} e^{z_k}}}
\end{eqnarray}

<t align="justify">

Donde $\mathbf{z} = [z_1, z_2, ..., z_K]$ es el vector de activación de la capa completamente conectada, $j$ es la clase objetivo y $K$ es el número total de clases. La función softmax asigna una probabilidad a cada clase, representando la confianza de la red en la predicción de esa clase.

Además de ajustar los pesos de las conexiones y los parámetros en la parte previa a la red completamente conectada, otro aspecto fundamental que se optimiza en una RNC es el valor de las entradas de la matriz kernel. El kernel es una matriz que se desliza sobre la imagen de entrada para realizar la operación de convolución, y sus valores determinan cómo se realiza la extracción de características en cada región.

Durante el entrenamiento de una RNC, los valores del kernel se optimizan junto con los demás parámetros de la red. El objetivo es encontrar los valores óptimos que permitan capturar características relevantes de la imagen y mejorar la capacidad de la red para realizar tareas de clasificación (*Ng, 2017*).

Esta optimización de los valores del kernel se realiza mediante el proceso de aprendizaje por retropropagación. Para actualizar los valores del kernel, se utiliza el algoritmo de SGD junto con el cálculo del gradiente de la función de pérdida con respecto a los valores del kernel.

Matemáticamente, la actualización de los valores del kernel se realiza de la siguiente manera:


\begin{eqnarray} 
  kernel_{nuevo} = kernel_{viejo} - \alpha \cdot \frac{\partial L}{\partial kernel_{viejo}}
\end{eqnarray}



<t align="justify">

Donde $kernel_{nuevo}$ representa los nuevos valores del kernel después de la actualización, $kernel_{viejo}$ son los valores actuales del kernel, $\alpha$ es la tasa de aprendizaje que controla la magnitud de la actualización y $\frac{\partial L}{\partial kernel_{viejo}}$ es el gradiente de la función de pérdida $L$ con respecto a los valores del kernel.

La optimización de los valores del kernel es crucial para mejorar el rendimiento de la red y su capacidad de generalización. Al ajustar estos valores, la RNC puede adaptarse mejor a los datos de entrenamiento y ser más sensible a los patrones y detalles importantes en las imágenes. Además, la optimización de los valores del kernel también puede ayudar a reducir el impacto del ruido y las variaciones irrelevantes en los datos, lo que mejora la robustez de la red (*Li, Zhang, & Liu, 2020*).

En resumen, la optimización de los valores del kernel en una RNC permite adaptar la red a los datos de entrenamiento, mejorar su capacidad de capturar características relevantes y aumentar su rendimiento en tareas de clasificación de imágenes.



***Entrenamiento de una red convolucional***

<t align="justify">

A continuación se describe la serie de pasos que conforma el entrenamiento de una RNC. 

1. Preparación de los datos:
   - Se recopilan y preparan los datos de entrenamiento, validación y prueba.
   - Se realiza el preprocesamiento de los datos, como normalización, reescalamiento o codificación.

2. Inicialización de los parámetros:
   - Se inicializan los pesos y sesgos de las capas de la RNC con valores aleatorios o preestablecidos.
   - Los parámetros de inicialización pueden influir en el rendimiento y la convergencia del modelo.

3. Propagación hacia adelante:
   - Se realiza la propagación hacia adelante de los datos de entrenamiento a través de las capas de la RNC.
   - Se calculan las salidas de cada capa utilizando las operaciones convolucionales, de pooling y de activación.

4. Cálculo de la función de pérdida:
   - Se calcula la función de pérdida, que evalúa la discrepancia entre las salidas predichas y las etiquetas de los datos de entrenamiento.
   - La elección de la función de pérdida depende del tipo de problema, como la pérdida de entropía cruzada para clasificación o la pérdida cuadrática media para regresión.

5. Retropropagación del error:
   - Se realiza la retropropagación del error desde la capa de salida hasta las capas anteriores de la RNC.
   - Se calcula el gradiente de la función de pérdida con respecto a los parámetros de la red, utilizando la regla de la cadena.

6. Actualización de los parámetros:
   - Se actualizan los pesos y sesgos de las capas de la RNC utilizando un algoritmo de optimización, como el SGD o Adam.
   - La actualización se realiza utilizando el gradiente calculado en el paso anterior y el coeficiente de aprendizaje.

7. Repetición de los pasos 3-6:
   - Se repiten los pasos 3-6 para cada lote de datos de entrenamiento, hasta agotar todos los lotes o alcanzar un número predeterminado de épocas.
   - El entrenamiento se realiza en lotes para aprovechar la eficiencia computacional y mejorar la estabilidad del modelo.

8. Evaluación y ajuste del modelo:
   - Se evalúa el rendimiento del modelo utilizando datos de validación o prueba, calculando métricas relevantes como precisión, exactitud o pérdida.
   - Si es necesario, se ajustan los hiperparámetros del modelo, como el tamaño del lote, el coeficiente de aprendizaje o el número de capas, para mejorar su rendimiento.

9. Predicción y despliegue del modelo:

#### Arquitectura de la Red Neuronal Utilizada

<t align="justify">

Se realizó una modificación al modelo VGG16 mediante una técnica llamada ajuste fino (fine-tuning). En primer lugar, se cargó el modelo VGG16 pre-entrenado sin las capas superiores (totalmente conectadas). Luego, se obtuvo el tensor de salida del modelo VGG16 y se aplicó una operación de promedio global para reducir las dimensiones espaciales del mapa de características.

Posteriormente, se agregó una capa completamente conectada con 128 unidades y función de activación ReLU. Se utilizó una capa de dropout con una tasa del 0.3 para reducir el sobreajuste. Finalmente, se agregó una capa completamente conectada final con 2 unidades para la clasificación binaria, utilizando la función de activación softmax.

Se creó un nuevo modelo combinando la entrada del modelo VGG16 y la salida de las capas completamente conectadas. Para el proceso de ajuste fino, se congelaron los pesos de las primeras 17 capas del modelo (hasta la última capa de max pooling), lo que significa que no se actualizarán durante el entrenamiento. En cambio, se permitió que los pesos de las capas restantes (desde la última capa de max pooling en adelante) se actualizaran durante el entrenamiento.

Esta modificación permite adaptar el modelo VGG16 pre-entrenado para el problema específico de clasificación binaria que se aborda en el proyecto. Al congelar parte de las capas pre-entrenadas y ajustar los pesos de las capas finales, se busca aprovechar el conocimiento aprendido en tareas anteriores para mejorar el rendimiento en la clasificación de las imágenes del proyecto.

```python
def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Matriz de confusión normalizada")
    else:
        print('Matriz de confusión, sin normalización')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('Clase real')
    plt.xlabel('Clase predicha')
    plt.tight_layout()

def plot_training_curves(history):
    acc = history.history['acc']
    val_acc = history.history['val_acc']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    
    epochs = range(1, len(acc) + 1)


    plt.title('Pérdida en entrenamiento y validacion')
    plt.plot(epochs, loss, 'bo', label='Pérdida en entrenamiento', color="r", linestyle = ":")
    plt.plot(epochs, val_loss, 'b', label='Pérdida en validación', color="g" )
    
    plt.legend()
    plt.figure()
    
    plt.title('Accuracy en entrenamiento y validación')
    plt.plot(epochs, acc, 'bo', label='Accuracy en entrenamiento', color="b", linestyle = ":")
    plt.plot(epochs, val_acc, 'b', label='Accuracy en validación', color="g")
    
    plt.legend()
    plt.figure()
    
    plt.show()

def print_evaluation(loss_value, accuracy_value):
    """
    Esta función recibe e imprime los valores de pérdida y de accuracy
    """
    print ('Pérdida: ', loss_value)
    print ('Accuracy: ', accuracy_value)

    
def print_results(cm):
    """
    Esta función recibe la matriz de confusión e imprime sus métricas asociadas
    """
    tp = cm[0][0] 
    tn = cm[1][1]
    fn = cm[0][1]
    fp = cm[1][0]
    precision = tp / (tp+fp)
    recall = tp / (tp+fn)
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)
    f1_score = ((2*precision*recall) / (recall+precision)) * 100
    print("Accuracy: %f \n Sensitivity : %f \n Specificity: %f \n F1 Score: %f" %(accuracy,sensitivity,specificity,f1_score))
    

def fine_tune_vgg():
    """
    Ajuste fino del modelo VGG16
    """
    # Cargar el modelo pre-entrenado VGG16 sin las capas superiores (totalmente conectadas)
    vgg16_model = VGG16(include_top=False, weights='imagenet')

    # Obtener el tensor de salida del modelo VGG16
    x = vgg16_model.output

    # Aplicar el promedio global para reducir las dimensiones espaciales del mapa de características
    x = GlobalAveragePooling2D()(x)

    # Agregar una capa completamente conectada con 128 unidades y función de activación ReLU
    x = Dense(128, activation='relu')(x)

    # Aplicar dropout con una tasa del 0.3 para reducir el sobreajuste
    x = Dropout(0.3)(x)

    # Agregar la capa completamente conectada final con 2 unidades (para clasificación binaria) y función de activación softmax
    predictions = Dense(2, activation='softmax')(x)

    # Crear un nuevo modelo usando la entrada del modelo VGG16 y la salida de las capas completamente conectadas
    model = Model(inputs=vgg16_model.input, outputs=predictions)

    # Congelar los pesos de las primeras 17 capas (hasta la última capa de max pooling)
    # Estas capas se establecen como no entrenables, por lo que sus pesos no se actualizarán durante el entrenamiento
    for layer in model.layers[:17]:
        layer.trainable = False

    # Permitir que los pesos de las capas restantes (desde la última capa de max pooling en adelante) se actualicen durante el entrenamiento
    for layer in model.layers[17:]:
        layer.trainable = True

    # Devolver el modelo compilado
    return model


def save_model(model, file_path):
    """
    Guarda el modelo en formato JSON en el archivo especificado. Recibe como entrada el modelo a guardar y la ruta del archivo.
    """
    model_json = model.to_json()
    with open('model.json', 'w') as json_file:
        json_file.write(model_json)

def train_network(model, class_weights, val_steps, callbacks, training_path, validation_path, val_batch_size, train_steps):
    """
    Entrena el modelo utilizando el generador de lotes de entrenamiento y validación.

    Args:
        model: El modelo a entrenar.
        class_weights: Pesos de clase utilizados para el cálculo de la función de pérdida durante el entrenamiento.
        val_steps: Número de pasos de validación por época.
        callbacks: Lista de objetos de devolución de llamada para personalizar el comportamiento durante el entrenamiento.
        training_path: Ruta del generador de lotes de entrenamiento.
        validation_path: Ruta del generador de lotes de validación.
        val_batch_size: Tamaño del lote de validación.
        train_steps: Número de pasos de entrenamiento por época.

    Returns:
        El objeto de historial resultante del entrenamiento del modelo.
    """

    return model.fit_generator(train_batches,
                              steps_per_epoch = train_steps,
                              class_weight = class_weights,
                              validation_data = valid_batches,
                              validation_steps = val_steps,
                              epochs = 1,
                              verbose = 1,
                              callbacks = callbacks)

def fine_tune_vgg16(train_batches, train_steps, class_weight, valid_batches, val_steps, file_path, callbacks):
    """
    Realiza el ajuste fino del modelo VGG16.

    Args:
        train_batches: Generador de lotes de entrenamiento.
        train_steps: Número de pasos de entrenamiento por época.
        class_weight: Pesos de clase utilizados para el cálculo de la función de pérdida durante el entrenamiento.
        valid_batches: Generador de lotes de validación.
        val_steps: Número de pasos de validación por época.
        file_path: Ruta del archivo donde se guardará el modelo después del ajuste fino.
        callbacks: Lista de objetos de devolución de llamada para personalizar el comportamiento durante el entrenamiento.

    Returns:
        El modelo después del ajuste fino y el objeto de historial resultante del entrenamiento.
    """
    # Cargar el modelo VGG16 pre-entrenado sin las capas superiores
    vgg16_model = VGG16(include_top=False, weights='imagenet')

    # Agregar capas convolucionales adicionales al modelo
    x = Conv2D(filters = 16, kernel_size = 3 , activation = 'relu', input_shape = (299, 299, 3))
    x = Conv2D(filters = 32, kernel_size = 3 , activation = 'relu')
    x = Conv2D(filters = 64, kernel_size = 3 , activation = 'relu')
    x = Conv2D(filters = 128, kernel_size = 3 , activation = 'relu')
    x = MaxPooling2D(pool_size = 3)
    x = vgg16_model.output

    # Agregar capas de regularización y clasificación al modelo
    x = Dropout(0.2)(x)
    x = Dense(128, activation='relu')(x)
    x = GlobalAveragePooling2D()(x)
    x = Dropout(0.2)(x)
    predictions = Dense(2, activation='softmax')(x)

    # Crear un nuevo modelo utilizando el modelo VGG16 pre-entrenado y las nuevas capas agregadas
    model = Model(inputs=vgg16_model.input, outputs=predictions)

    # Compilar el modelo para el entrenamiento
    model.compile(Adam(lr=0.000095), loss='categorical_crossentropy', metrics=['accuracy'])

    # Entrenar el modelo sin ajuste fino
    model.fit_generator(train_batches,
                              steps_per_epoch = train_steps,
                              class_weight = class_weights,
                              validation_data = valid_batches,
                              validation_steps = val_steps,
                              epochs = 20,
                              verbose = 1,
                              callbacks = callbacks)

    # Congelar los pesos de las primeras 17 capas (hasta la última capa de max pooling)
    for layer in model.layers[:17]:
        layer.trainable = False

    # Permitir que los pesos de las capas restantes (desde la última capa de max pooling en adelante) se actualicen durante el ajuste fino
    for layer in model.layers[17:]:
        layer.trainable = True

    # Compilar nuevamente el modelo para el ajuste fino
    model.compile(Adam(lr=0.000095), loss='categorical_crossentropy', metrics=['accuracy'])

    # Entrenar el modelo con el ajuste fino
    history = model.fit_generator(train_batches,
                              steps_per_epoch = train_steps,
                              class_weight = class_weights,
                              validation_data = valid_batches,
                              validation_steps = val_steps,
                              epochs = 20,
                              verbose = 1,
                              callbacks = callbacks)
    return model, history

    
    
    
```

#### Callbacks y Entrenamiento de la Red Neuronal

<t align="justify">

Los callbacks desempeñan un papel fundamental en el proceso de entrenamiento de modelos de aprendizaje automático, ya que permiten mejorar y controlar el rendimiento del modelo durante el entrenamiento. Estos objetos, integrados en las bibliotecas de TensorFlow y Keras, ofrecen funcionalidades clave, como guardar los mejores modelos, ajustar la tasa de aprendizaje y detener el entrenamiento anticipadamente si es necesario.

Los parámetros definidos en los callbacks tienen un papel clave en el proceso de entrenamiento del modelo. Veamos la importancia de cada uno de ellos:

- *ModelCheckpoint:* Este nos permite especificar la ruta donde se guardará el modelo así como la manera en que la red identificará al mejor modelo.

- *ReduceLROnPlateau:* Esta función permite monitorear y adecuar el proceso de entrenamiento de la red neuronal, mediante el ajuste de la tasa de aprendizaje. 

- *EarlyStopping:* El parámetro de EarlyStopping desempeña un papel crucial al permitir especificar cuándo detener el entrenamiento de la red neuronal. Mediante la configuración de este parámetro, determinamos la métrica utilizada para evaluar el rendimiento, el umbral mínimo de mejora en la métrica que se considera significativo y el número de épocas sin mejora que se tolerará antes de detener el entrenamiento (patience).

Durante el desarrollo del proyecto, realizamos diferentes entrenamientos ajustando los parámetros de EarlyStopping. Al aumentar la paciencia (patience) y reducir el umbral mínimo de mejora, la red requerirá un mayor tiempo de entrenamiento. Sin embargo, esperaríamos obtener resultados más precisos y adecuados.

```python
# Definimos los callbacks para mejorar y controlar el proceso de entrenamiento del modelo
callbacks = [
        # Guarda el modelo con la mejor precisión en el conjunto de validación
        ModelCheckpoint(file_path, monitor = 'val_acc', verbose = 1, save_best_only = True, mode = 'max'),

        # Reduce la tasa de aprendizaje si no hay mejora en la precisión en el conjunto de validación
        ReduceLROnPlateau(monitor = 'val_acc', factor = 0.5, patience = 10, verbose = 1, mode = 'max', min_lr = 0.0000314159265359),

        # Detiene el entrenamiento anticipadamente si no hay mejora en la pérdida en el conjunto de validación
        EarlyStopping(monitor = 'val_loss', min_delta = 1e-2, patience = 3, verbose = 1)
        ]

# Realiza el ajuste fino del modelo VGG16
model, history = fine_tune_vgg16(train_batches, train_steps, class_weights, valid_batches, val_steps, file_path, callbacks)
# Guarda el modelo utilizando la función save_model
save_model(model, file_path)
```

```python
# Asigna las etiquetas de clase correspondientes a los datos de prueba: Melanoma y No melanoma
test_labels = test_batches.classes

# Realiza predicciones utilizando el modelo entrenado
predictions = model.predict_generator(test_batches, steps = val_steps, verbose = 1)

# Imprime la matriz de confusión
cm = confusion_matrix(test_labels, predictions.argmax(axis=1))

# Imprime las métricas asociadas a la matriz de confusión para evaluar el desempeño del modelo entrenado
print_results(cm)

# Grafica la matriz de confusión
plot_confusion_matrix(cm, ['Melanoma', 'No melanoma'])
```
