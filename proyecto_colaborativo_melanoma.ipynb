***Convolucin y activacin (ReLU) en redes neuronales convolucionales***

<t align="justify">

Las RNC se caracterizan por su capacidad para procesar imgenes de manera eficiente y capturar caractersticas importantes mediante el uso de convoluciones y funciones de activacin (*LeCun, Bengio, & Hinton, 2015*).

La convolucin es una operacin fundamental en las RNC que permite capturar patrones locales en las imgenes. En esta operacin, se utiliza un kernel, que es una pequea ventana deslizante, para realizar una multiplicacin elemento por elemento entre los valores de los pxeles de la imagen y los valores del kernel (*Goodfellow, Bengio, & Courville, 2016*). Esto se puede expresar matemticamente de la siguiente manera:



\begin{eqnarray} 
  \mathbf{C}(i, j) = \sum_{m}\sum_{n} \mathbf{I}(i+m, j+n) \cdot \mathbf{K}(m, n)
\end{eqnarray}


<t align="justify">

donde $\mathbf{C}(i, j)$ es el valor resultante de la convolucin en la posicin $(i, j)$, $\mathbf{I}(i+m, j+n)$ es el valor del pxel de la imagen en la posicin $(i+m, j+n)$, y $\mathbf{K}(m, n)$ es el valor del kernel en la posicin $(m, n)$.

La convolucin se aplica en toda la imagen deslizando el kernel sobre ella, generando as una representacin convolucionada de la imagen que resalta caractersticas locales. Esta operacin permite a la red neuronal convolucional detectar bordes, texturas y otros patrones visuales importantes en diferentes regiones de la imagen.

Despus de la convolucin, se aplica la funcin de activacin Rectificadora linear (o ReLU por sus siglas en ingls) para introducir no linealidad en la red (*Krizhevsky et al., 2012*). La funcin ReLU se define de la siguiente manera:


\begin{eqnarray} 
f(x) = \max(0, x)
\end{eqnarray}


<t align="justify">

donde $x$ es el valor de entrada. La funcin ReLU mapea los valores negativos a cero y mantiene los valores positivos sin cambios. Esta funcin de activacin juega un papel crucial en la capacidad de la red para aprender representaciones ms complejas y discriminativas (*Krizhevsky et al., 2012*).

La combinacin de la convolucin y la activacin ReLU permite a las RNC capturar y resaltar caractersticas importantes en las imgenes de manera eficiente y no lineal (*LeCun et al., 2015*).

***Pooling en Redes Neuronales Convolucionales***

<t align="justify">

Despus de la etapa de convolucin y activacin en las RNC, se aplica una operacin conocida como "pooling". El pooling es una tcnica que reduce la dimensionalidad espacial de la representacin convolucionada, preservando las caractersticas ms importantes (*Krizhevsky, Sutskever, & Hinton, 2012*).

El tipo ms comn de pooling es el max pooling, donde se divide la imagen en regiones no superpuestas y se toma el valor mximo de cada regin. Esta operacin se puede expresar de la siguiente manera:


\begin{eqnarray}

\end{eqnarray}

<t align="justify">

donde $\mathbf{P}(i, j)$ es el valor resultante del pooling en la posicin $(i, j)$, $\mathbf{C}(i \cdot s + m, j \cdot s + n)$ es el valor en la representacin convolucionada en la posicin $(i \cdot s + m, j \cdot s + n)$, y $s$ es el tamao del pool o filtro (*Krizhevsky et al., 2012*).

El pooling se utiliza para reducir el tamao espacial de la representacin, lo que a su vez reduce la cantidad de parmetros y operaciones en la red. Adems, el pooling proporciona invariancia a pequeas traslaciones y deformaciones en las caractersticas detectadas, lo que puede mejorar la capacidad de generalizacin de la red (*Krizhevsky et al., 2012*).

Otro tipo de pooling comnmente utilizado es el average pooling, que toma el promedio de los valores en lugar del mximo. Sin embargo, el max pooling ha demostrado ser ms efectivo en la prctica, ya que preserva las caractersticas ms relevantes y proporciona una mejor invariancia a pequeas perturbaciones (*Krizhevsky et al., 2012*).

El proceso de pooling se repite en varias capas de la red, reduciendo gradualmente la dimensionalidad espacial y extrayendo caractersticas cada vez ms abstractas. Esto permite a la red focalizarse en las caractersticas ms importantes y generalizar mejor a diferentes instancias de una misma clase (*Krizhevsky et al., 2012*).



***Reduccin de dimensionalidad***

<t align="justify">

Despus del proceso de pooling, la representacin resultante se somete a una reduccin de dimensionalidad en las RNC. Esta etapa tiene como objetivo transformar la representacin multidimensional obtenida en una forma ms manejable y adecuada para la clasificacin final.

\end{eqnarray}


Donde la funcin $\text{aplanar}(\cdot)$ convierte la representacin multidimensional en un vector lineal (*Goodfellow et al., 2016*).

Despus de la reduccin de dimensionalidad, la representacin unidimensional resultante se alimenta a una capa completamente conectada en las RNC. Esta capa tiene como objetivo finalizar el proceso de clasificacin y tomar decisiones basadas en la informacin extrada de las capas anteriores.

***Optimizacin de la red neuronal convolucional***

<t align="justify">

La optimizacin de una red neuronal es un proceso fundamental que busca ajustar los parmetros de la red para que pueda realizar una clasificacin precisa. En el caso de las RNC, se optimizan los parmetros tanto en la parte previa a la red completamente conectada como en la propia red completamente conectada.

En la parte previa a la red completamente conectada, se utilizan tcnicas de optimizacin para ajustar los pesos de las conexiones en las capas convolucionales y de pooling. Un mtodo popularmente utilizado en esta etapa es el algoritmo Adam (*Kingma & Ba, 2015*). Adam combina las ventajas del algoritmo de descenso de gradiente estocstico (o SGD por sus siglas en ingls) con una adaptacin del coeficiente de aprendizaje en funcin del momento y la estimacin del segundo momento del gradiente.

Matemticamente, el algoritmo Adam actualiza los pesos ww de la red en cada iteracin utilizando las siguientes ecuaciones:


\begin{eqnarray} 
  m = \beta_1 \cdot m + (1 - \beta_1) \cdot \nabla_w J(w)
\end{eqnarray}

\begin{eqnarray} 
  v = \beta_2 \cdot v + (1 - \beta_2) \cdot (\nabla_w J(w))^2
\end{eqnarray}

\begin{eqnarray} 
  \hat{m} = \frac{m}{1 - \beta_1^t}
\end{eqnarray}
 
\begin{eqnarray} 
  \hat{v} = \frac{v}{1 - \beta_2^t}
\end{eqnarray}

\begin{eqnarray} 
  w = w - \alpha \cdot \frac{\hat{m}}{\sqrt{\hat{v}} + \epsilon}
\end{eqnarray}


<t align="justify">

Donde $\nabla_w$ es el gradiente de la funcin de prdida $J(w)$ con respecto a los pesos $w$, $\beta_1$ y $\beta_2$ son hiperparmetros que controlan la ponderacin del momento y el segundo momento del gradiente, respectivamente, $m$ y $v$ son estimaciones del primer momento (media) y segundo momento (varianza) del gradiente, $t$ es el contador de iteraciones, $\hat{m}$ y $\hat{v}$ son las versiones corregidas de los momentos, $\alpha$ es el coeficiente de aprendizaje y $\epsilon$ es una pequea constante para evitar divisiones por cero.

En la capa completamente conectada de la RNC, se utiliza la funcin de activacin softmax para obtener una distribucin de probabilidad sobre las posibles clases de salida. La funcin softmax se define matemticamente de la siguiente manera:


\begin{eqnarray} 
  \text{softmax}(\mathbf{z})_j = \frac{{e^{z_j}}}{{\sum_{k=1}^{K} e^{z_k}}}
\end{eqnarray}

<t align="justify">

Donde $\mathbf{z} = [z_1, z_2, ..., z_K]$ es el vector de activacin de la capa completamente conectada, $j$ es la clase objetivo y $K$ es el nmero total de clases. La funcin softmax asigna una probabilidad a cada clase, representando la confianza de la red en la prediccin de esa clase.

Adems de ajustar los pesos de las conexiones y los parmetros en la parte previa a la red completamente conectada, otro aspecto fundamental que se optimiza en una RNC es el valor de las entradas de la matriz kernel. El kernel es una matriz que se desliza sobre la imagen de entrada para realizar la operacin de convolucin, y sus valores determinan cmo se realiza la extraccin de caractersticas en cada regin.

Durante el entrenamiento de una RNC, los valores del kernel se optimizan junto con los dems parmetros de la red. El objetivo es encontrar los valores ptimos que permitan capturar caractersticas relevantes de la imagen y mejorar la capacidad de la red para realizar tareas de clasificacin (*Ng, 2017*).

Esta optimizacin de los valores del kernel se realiza mediante el proceso de aprendizaje por retropropagacin. Para actualizar los valores del kernel, se utiliza el algoritmo de SGD junto con el clculo del gradiente de la funcin de prdida con respecto a los valores del kernel.

Matemticamente, la actualizacin de los valores del kernel se realiza de la siguiente manera:


\begin{eqnarray} 
  kernel_{nuevo} = kernel_{viejo} - \alpha \cdot \frac{\partial L}{\partial kernel_{viejo}}
\end{eqnarray}



<t align="justify">

Donde $kernel_{nuevo}$ representa los nuevos valores del kernel despus de la actualizacin, $kernel_{viejo}$ son los valores actuales del kernel, $\alpha$ es la tasa de aprendizaje que controla la magnitud de la actualizacin y $\frac{\partial L}{\partial kernel_{viejo}}$ es el gradiente de la funcin de prdida $L$ con respecto a los valores del kernel.

La optimizacin de los valores del kernel es crucial para mejorar el rendimiento de la red y su capacidad de generalizacin. Al ajustar estos valores, la RNC puede adaptarse mejor a los datos de entrenamiento y ser ms sensible a los patrones y detalles importantes en las imgenes. Adems, la optimizacin de los valores del kernel tambin puede ayudar a reducir el impacto del ruido y las variaciones irrelevantes en los datos, lo que mejora la robustez de la red (*Li, Zhang, & Liu, 2020*).

En resumen, la optimizacin de los valores del kernel en una RNC permite adaptar la red a los datos de entrenamiento, mejorar su capacidad de capturar caractersticas relevantes y aumentar su rendimiento en tareas de clasificacin de imgenes.

3. Propagacin hacia adelante:

8. Evaluacin y ajuste del modelo:

9. Prediccin y despliegue del modelo:

#### Arquitectura de la Red Neuronal Utilizada

<t align="justify">

Se realiz una modificacin al modelo VGG16 mediante una tcnica llamada ajuste fino (fine-tuning). En primer lugar, se carg el modelo VGG16 pre-entrenado sin las capas superiores (totalmente conectadas). Luego, se obtuvo el tensor de salida del modelo VGG16 y se aplic una operacin de promedio global para reducir las dimensiones espaciales del mapa de caractersticas.

Posteriormente, se agreg una capa completamente conectada con 128 unidades y funcin de activacin ReLU. Se utiliz una capa de dropout con una tasa del 0.3 para reducir el sobreajuste. Finalmente, se agreg una capa completamente conectada final con 2 unidades para la clasificacin binaria, utilizando la funcin de activacin softmax.

Se cre un nuevo modelo combinando la entrada del modelo VGG16 y la salida de las capas completamente conectadas. Para el proceso de ajuste fino, se congelaron los pesos de las primeras 17 capas del modelo (hasta la ltima capa de max pooling), lo que significa que no se actualizarn durante el entrenamiento. En cambio, se permiti que los pesos de las capas restantes (desde la ltima capa de max pooling en adelante) se actualizaran durante el entrenamiento.

Esta modificacin permite adaptar el modelo VGG16 pre-entrenado para el problema especfico de clasificacin binaria que se aborda en el proyecto. Al congelar parte de las capas pre-entrenadas y ajustar los pesos de las capas finales, se busca aprovechar el conocimiento aprendido en tareas anteriores para mejorar el rendimiento en la clasificacin de las imgenes del proyecto.

```python
def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Matriz de confusin normalizada")
    else:
        print('Matriz de confusin, sin normalizacin')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('Clase real')
    plt.xlabel('Clase predicha')
    plt.tight_layout()

def plot_training_curves(history):
    acc = history.history['acc']
    val_acc = history.history['val_acc']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    
    epochs = range(1, len(acc) + 1)


    plt.title('Prdida en entrenamiento y validacion')
    plt.plot(epochs, loss, 'bo', label='Prdida en entrenamiento', color="r", linestyle = ":")
    plt.plot(epochs, val_loss, 'b', label='Prdida en validacin', color="g" )
    
    plt.legend()
    plt.figure()
    
    plt.title('Accuracy en entrenamiento y validacin')
    plt.plot(epochs, acc, 'bo', label='Accuracy en entrenamiento', color="b", linestyle = ":")
    plt.plot(epochs, val_acc, 'b', label='Accuracy en validacin', color="g")
    
    plt.legend()
    plt.figure()
    
    plt.show()

def print_evaluation(loss_value, accuracy_value):
    """
    Esta funcin recibe e imprime los valores de prdida y de accuracy
    """
    print ('Prdida: ', loss_value)
    print ('Accuracy: ', accuracy_value)

    
def print_results(cm):
    """
    Esta funcin recibe la matriz de confusin e imprime sus mtricas asociadas
    """
    tp = cm[0][0] 
    tn = cm[1][1]
    fn = cm[0][1]
    fp = cm[1][0]
    precision = tp / (tp+fp)
    recall = tp / (tp+fn)
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)
    f1_score = ((2*precision*recall) / (recall+precision)) * 100
    print("Accuracy: %f \n Sensitivity : %f \n Specificity: %f \n F1 Score: %f" %(accuracy,sensitivity,specificity,f1_score))
    

def fine_tune_vgg():
    """
    Ajuste fino del modelo VGG16
    """
    # Cargar el modelo pre-entrenado VGG16 sin las capas superiores (totalmente conectadas)
    vgg16_model = VGG16(include_top=False, weights='imagenet')

    # Obtener el tensor de salida del modelo VGG16
    x = vgg16_model.output

    # Aplicar el promedio global para reducir las dimensiones espaciales del mapa de caractersticas
    x = GlobalAveragePooling2D()(x)

    # Agregar una capa completamente conectada con 128 unidades y funcin de activacin ReLU
    x = Dense(128, activation='relu')(x)

    # Aplicar dropout con una tasa del 0.3 para reducir el sobreajuste
    x = Dropout(0.3)(x)

    # Agregar la capa completamente conectada final con 2 unidades (para clasificacin binaria) y funcin de activacin softmax
    predictions = Dense(2, activation='softmax')(x)

    # Crear un nuevo modelo usando la entrada del modelo VGG16 y la salida de las capas completamente conectadas
    model = Model(inputs=vgg16_model.input, outputs=predictions)

    # Congelar los pesos de las primeras 17 capas (hasta la ltima capa de max pooling)
    # Estas capas se establecen como no entrenables, por lo que sus pesos no se actualizarn durante el entrenamiento
    for layer in model.layers[:17]:
        layer.trainable = False

    # Permitir que los pesos de las capas restantes (desde la ltima capa de max pooling en adelante) se actualicen durante el entrenamiento
    for layer in model.layers[17:]:
        layer.trainable = True

    # Devolver el modelo compilado
    return model


def save_model(model, file_path):
    """
    Guarda el modelo en formato JSON en el archivo especificado. Recibe como entrada el modelo a guardar y la ruta del archivo.
    """
    model_json = model.to_json()
    with open('model.json', 'w') as json_file:
        json_file.write(model_json)

def train_network(model, class_weights, val_steps, callbacks, training_path, validation_path, val_batch_size, train_steps):
    """
    Entrena el modelo utilizando el generador de lotes de entrenamiento y validacin.

    Args:
        model: El modelo a entrenar.
        class_weights: Pesos de clase utilizados para el clculo de la funcin de prdida durante el entrenamiento.
        val_steps: Nmero de pasos de validacin por poca.
        callbacks: Lista de objetos de devolucin de llamada para personalizar el comportamiento durante el entrenamiento.
        training_path: Ruta del generador de lotes de entrenamiento.
        validation_path: Ruta del generador de lotes de validacin.
        val_batch_size: Tamao del lote de validacin.
        train_steps: Nmero de pasos de entrenamiento por poca.

    Returns:
        El objeto de historial resultante del entrenamiento del modelo.
    """

    return model.fit_generator(train_batches,
                              steps_per_epoch = train_steps,
                              class_weight = class_weights,
                              validation_data = valid_batches,
                              validation_steps = val_steps,
                              epochs = 1,
                              verbose = 1,
                              callbacks = callbacks)

def fine_tune_vgg16(train_batches, train_steps, class_weight, valid_batches, val_steps, file_path, callbacks):
    """
    Realiza el ajuste fino del modelo VGG16.

    Args:
        train_batches: Generador de lotes de entrenamiento.
        train_steps: Nmero de pasos de entrenamiento por poca.
        class_weight: Pesos de clase utilizados para el clculo de la funcin de prdida durante el entrenamiento.
        valid_batches: Generador de lotes de validacin.
        val_steps: Nmero de pasos de validacin por poca.
        file_path: Ruta del archivo donde se guardar el modelo despus del ajuste fino.
        callbacks: Lista de objetos de devolucin de llamada para personalizar el comportamiento durante el entrenamiento.

    Returns:
        El modelo despus del ajuste fino y el objeto de historial resultante del entrenamiento.
    """
    # Cargar el modelo VGG16 pre-entrenado sin las capas superiores
    vgg16_model = VGG16(include_top=False, weights='imagenet')

    # Agregar capas convolucionales adicionales al modelo
    x = Conv2D(filters = 16, kernel_size = 3 , activation = 'relu', input_shape = (299, 299, 3))
    x = Conv2D(filters = 32, kernel_size = 3 , activation = 'relu')
    x = Conv2D(filters = 64, kernel_size = 3 , activation = 'relu')
    x = Conv2D(filters = 128, kernel_size = 3 , activation = 'relu')
    x = MaxPooling2D(pool_size = 3)
    x = vgg16_model.output

    # Agregar capas de regularizacin y clasificacin al modelo
    x = Dropout(0.2)(x)
    x = Dense(128, activation='relu')(x)
    x = GlobalAveragePooling2D()(x)
    x = Dropout(0.2)(x)
    predictions = Dense(2, activation='softmax')(x)

    # Crear un nuevo modelo utilizando el modelo VGG16 pre-entrenado y las nuevas capas agregadas
    model = Model(inputs=vgg16_model.input, outputs=predictions)

    # Compilar el modelo para el entrenamiento
    model.compile(Adam(lr=0.000095), loss='categorical_crossentropy', metrics=['accuracy'])

    # Entrenar el modelo sin ajuste fino
    model.fit_generator(train_batches,
                              steps_per_epoch = train_steps,
                              class_weight = class_weights,
                              validation_data = valid_batches,
                              validation_steps = val_steps,
                              epochs = 20,
                              verbose = 1,
                              callbacks = callbacks)

    # Congelar los pesos de las primeras 17 capas (hasta la ltima capa de max pooling)
    for layer in model.layers[:17]:
        layer.trainable = False

    # Permitir que los pesos de las capas restantes (desde la ltima capa de max pooling en adelante) se actualicen durante el ajuste fino
    for layer in model.layers[17:]:
        layer.trainable = True

    # Compilar nuevamente el modelo para el ajuste fino
    model.compile(Adam(lr=0.000095), loss='categorical_crossentropy', metrics=['accuracy'])

    # Entrenar el modelo con el ajuste fino
    history = model.fit_generator(train_batches,
                              steps_per_epoch = train_steps,
                              class_weight = class_weights,
                              validation_data = valid_batches,
                              validation_steps = val_steps,
                              epochs = 20,
                              verbose = 1,
                              callbacks = callbacks)
    return model, history

    
    
    
```

```python
# Directorio donde guardamos el modelo entrenado
file_path = 'weights'

# Directorio donde se encuentran los distintos datos ya divididos
training_path = '/kaggle/input/melanoma/dermmel/DermMel/train_sep'
validation_path = '/kaggle/input/melanoma/dermmel/DermMel/valid'
test_path = '/kaggle/input/melanoma/dermmel/DermMel/test'

# Cantidad de datos en cada conjunto
num_train_samples = 10682
num_val_samples = 3562
num_test_samples = 3562

# Tamaos de batch para el entrenamiento: 8, 16, 32, 64, 128
train_batch_size = 16
val_batch_size = 16
test_batch_size = 16

# Calcula el nmero de pasos de entrenamiento necesarios para cubrir todos las imgenes de entrenamiento
train_steps = np.ceil(num_train_samples / train_batch_size)
val_steps = np.ceil(num_val_samples / val_batch_size)
test_steps = np.ceil(num_val_samples / val_batch_size)

# Pesos de castigo y reforzamiento para cada categora
class_weights = {
        0: 5.1, # melanoma
        1: 1.0 # non-melanoma
}

# Preprocesamiento de las imgenes para ajustar tamao y reescalar los valores de cada pixel a [0,1]
# Entrenamiento
train_batches = ImageDataGenerator(rescale = 1./255).flow_from_directory(training_path,
                                    target_size = (299, 299),
                                    batch_size = val_batch_size,
                                    class_mode = 'categorical')

# Validacin
valid_batches = ImageDataGenerator(rescale = 1./255).flow_from_directory(validation_path,
                                    target_size = (299, 299),
                                    batch_size = val_batch_size,
                                    class_mode = 'categorical')

# Prue
test_batches = ImageDataGenerator(rescale = 1./255).flow_from_directory(test_path,
                                    target_size = (299, 299),
                                    batch_size = test_batch_size,
                                    class_mode = 'categorical',
                                    shuffle = False)
                                    
```

#### Callbacks y Entrenamiento de la Red Neuronal

<t align="justify">

Los callbacks desempean un papel fundamental en el proceso de entrenamiento de modelos de aprendizaje automtico, ya que permiten mejorar y controlar el rendimiento del modelo durante el entrenamiento. Estos objetos, integrados en las bibliotecas de TensorFlow y Keras, ofrecen funcionalidades clave, como guardar los mejores modelos, ajustar la tasa de aprendizaje y detener el entrenamiento anticipadamente si es necesario.

Los parmetros definidos en los callbacks tienen un papel clave en el proceso de entrenamiento del modelo. Veamos la importancia de cada uno de ellos:

- *ModelCheckpoint:* Este nos permite especificar la ruta donde se guardar el modelo as como la manera en que la red identificar al mejor modelo.

- *ReduceLROnPlateau:* Esta funcin permite monitorear y adecuar el proceso de entrenamiento de la red neuronal, mediante el ajuste de la tasa de aprendizaje. 

- *EarlyStopping:* El parmetro de EarlyStopping desempea un papel crucial al permitir especificar cundo detener el entrenamiento de la red neuronal. Mediante la configuracin de este parmetro, determinamos la mtrica utilizada para evaluar el rendimiento, el umbral mnimo de mejora en la mtrica que se considera significativo y el nmero de pocas sin mejora que se tolerar antes de detener el entrenamiento (patience).

Durante el desarrollo del proyecto, realizamos diferentes entrenamientos ajustando los parmetros de EarlyStopping. Al aumentar la paciencia (patience) y reducir el umbral mnimo de mejora, la red requerir un mayor tiempo de entrenamiento. Sin embargo, esperaramos obtener resultados ms precisos y adecuados.

```python
# Definimos los callbacks para mejorar y controlar el proceso de entrenamiento del modelo
callbacks = [
        # Guarda el modelo con la mejor precisin en el conjunto de validacin
        ModelCheckpoint(file_path, monitor = 'val_acc', verbose = 1, save_best_only = True, mode = 'max'),

        # Reduce la tasa de aprendizaje si no hay mejora en la precisin en el conjunto de validacin
        ReduceLROnPlateau(monitor = 'val_acc', factor = 0.5, patience = 10, verbose = 1, mode = 'max', min_lr = 0.0000314159265359),

        # Detiene el entrenamiento anticipadamente si no hay mejora en la prdida en el conjunto de validacin
        EarlyStopping(monitor = 'val_loss', min_delta = 1e-2, patience = 3, verbose = 1)
        ]

# Realiza el ajuste fino del modelo VGG16
model, history = fine_tune_vgg16(train_batches, train_steps, class_weights, valid_batches, val_steps, file_path, callbacks)
# Guarda el modelo utilizando la funcin save_model
save_model(model, file_path)
```

```python
# Asigna las etiquetas de clase correspondientes a los datos de prueba: Melanoma y No melanoma
test_labels = test_batches.classes

# Realiza predicciones utilizando el modelo entrenado
predictions = model.predict_generator(test_batches, steps = val_steps, verbose = 1)

# Imprime la matriz de confusin
cm = confusion_matrix(test_labels, predictions.argmax(axis=1))

# Imprime las mtricas asociadas a la matriz de confusin para evaluar el desempeo del modelo entrenado
print_results(cm)

# Grafica la matriz de confusin
plot_confusion_matrix(cm, ['Melanoma', 'No melanoma'])
```
